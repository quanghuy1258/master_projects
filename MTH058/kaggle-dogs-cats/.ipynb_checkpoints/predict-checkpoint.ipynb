```python
import sys, os
import numpy as np

import matplotlib.pyplot as plt

import common
import read_dataset
```

```python
# Read data
X_test, y_test = read_dataset.read_data('test_new/')
```

```python
# Load model
model = common.create_cnn_model()

# Load weights
checkpoint_filepath = 'training/cnn_9227_128_8l_100.h5'
try:
    model.load_weights(checkpoint_filepath)
except IndexError:
    print('Error in loading weights')
```

```python
label_map = {1.0: "cat", 0.0: "dog"}
```

```python
def predict_and_show(index_image):
    image = X_test[index_image]
    correct_label = label_map[y_test[index_image]]

    pred = model.predict_classes(image.reshape(1, 128, 128, 3))
    predicted_label = label_map[pred[0][0]]

    plt.imshow(image);
    plt.title(f'Correct label: {correct_label} | Predicted label: {predicted_label}');
```

```python
# Show iamge
index_image = int(input('Enter index of image in dataset: '))
predict_and_show(index_image)
```
